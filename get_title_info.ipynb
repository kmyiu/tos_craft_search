```python
import pandas as pd
import numpy as np
import re
```

```python
titlestxt = open(r'titleslist.txt','r', encoding='UTF-8')
```

```python
x = titlestxt.readlines()
```

```python
df = pd.DataFrame()
props = ['lv', 'name', 'cond', 'type']
```

```python
for row in x:
    print(row)
    lv = row.split("|")[1]
    name = row.split("|")[2]
    cond = row.split("|")[3].split("=")[1].split("}")[0]
    titletype = ""
    title = pd.Series([lv, name, cond, titletype], index = props)
    df = df.append(title, ignore_index=True)
```

```python
df.to_csv('titles.csv', encoding='utf-8',index=False)
```
